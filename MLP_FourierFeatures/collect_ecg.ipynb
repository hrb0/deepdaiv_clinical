```python
import pandas as pd
import requests
import os

# CSV 파일 로드
csv_path = 'C:/Users/yumi/projects/deepdaiv_clinical/mimic_preprocessed/MI_F.csv'
df = pd.read_csv(csv_path)

# 파일이 저장될 기본 경로 설정
base_download_path = 'D:/mimiciv_data/mimic-iv-ecg/1.0/files'

for index, row in df.iterrows():
    subject_id = row['subject_id']
    ecg_study_id = row['ecg_study_id']
    
    # 파일 다운로드 URL과 저장 경로 생성
    subject_folder = f"p{str(subject_id)[:4]}/p{subject_id}"
    study_folder = f"s{ecg_study_id}"
    file_base_url = f"https://physionet.org/files/mimic-iv-ecg/1.0/files/{subject_folder}/{study_folder}/"
    download_path = os.path.join(base_download_path, subject_folder, study_folder)
    
    # 다운로드할 파일이 있는 경우
    if not os.path.exists(download_path):
        os.makedirs(download_path)
    
    try:
        # .dat 및 .hea 파일 다운로드
        for file_ext in ['.dat', '.hea']:
            file_url = f"{file_base_url}{ecg_study_id}{file_ext}"
            response = requests.get(file_url)
            # 요청이 성공했을 경우에만 파일 저장
            if response.status_code == 200:
                with open(os.path.join(download_path, f"{ecg_study_id}{file_ext}"), 'wb') as f:
                    f.write(response.content)
                print(f"{ecg_study_id}{file_ext} 다운로드 완료.")
            else:
                print(f"{ecg_study_id}{file_ext} 파일을 찾을 수 없습니다.")
    except Exception as e:
        print(f"다운로드 중 오류 발생: {e}")
```
